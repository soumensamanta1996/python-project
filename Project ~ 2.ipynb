```python
##replace all outlier values for numerical value by mean

#ans:2.2
import numpy as np
    
def replace_outliers(df):
    
    q1=df.quantile(0.25)

    q3=df.quantile(0.75)

    IQR=q3-q1

    
    upper = df[~(df>(q3+1.5*IQR))].max()
    
    lower = df[~(df<(q1-1.5*IQR))].min()

    df = np.where(df > upper, df.mean(),np.where(df < lower,df.mean(),df))

    return df  
```

```python
replace_outliers(df1[ 'driving licence present'])
```

```python
replace_outliers(df2['annual premium'])
replace_outliers(df2['response'])
```
